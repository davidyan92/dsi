```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_predict, cross_val_score, KFold
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from scipy import stats
import patsy

%config InlineBackend.figure_format ='retina'
%matplotlib inline
```

```python
# read the file into dataframe
file_path = './datasets/discrim.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# get highest frequency value for prices
psoda_hf = df['psoda'].value_counts().sort_values(ascending=False).head(1).values[0]
pfries_hf = df['pfries'].value_counts().sort_values(ascending=False).head(1).values[0]
pentree_hf = df['pentree'].value_counts().sort_values(ascending=False).head(1).values[0]
```

```python
# fix null rows for prices
df['psoda'].fillna(psoda_hf, inplace=True)
df['pfries'].fillna(pfries_hf, inplace=True)
df['pentree'].fillna(pentree_hf, inplace=True)
```

```python
# fix predictor columns
df[df['prpblck'].isnull()]
```

```python
# since its just 1 row, we remove it
new_df = df.drop(labels=384)
```

```python
# get all required columns
# prpblck       proportion black, zipcode
# prppov        proportion in poverty, zipcode
# prpncar       proportion no car, zipcode
final_df = new_df[['psoda','pfries','pentree','prpblck','prppov','prpncar','chain']]
```

```python
final_df.describe()
```

```python
# remove outliers from price columns
final_final_df = final_df[(np.abs(stats.zscore(final_df[['psoda','pfries','pentree']])) < 3).all(axis=1)]
```

```python
# Investigate outliers for prices
prices = final_final_df[['psoda','pfries','pentree']]
sns.boxplot(data=prices)
```

```python
# feature engineering to get price for 1 meal
final_final_df['pmeal'] = final_final_df['psoda'] + final_final_df['pfries'] + final_final_df['pentree']
```

```python
# EDA for each food type
sns.pairplot(final_final_df[['pmeal','prpblck','prppov','prpncar']])
```

```python
BK = final_final_df[final_final_df['chain']==1]
KFC = final_final_df[final_final_df['chain']==2]
RR = final_final_df[final_final_df['chain']==3]
WEN = final_final_df[final_final_df['chain']==4]
```

```python
sns.pairplot(KFC[['pmeal','prpblck','prppov','prpncar']])
```

```python
# Null hypothesis will be that there will be no significant predictions between prpblck, prppov, prpncar with price of meal
# meaning all slope coefficients are 0 or p-value above 0.05
# We test for all restaurants first
# chain BK = 1, KFC = 2, Roy Rogers = 3, Wendy's = 4
formula = 'pmeal ~ prpblck + prpncar'
y, X= patsy.dmatrices(formula, data=WEN, return_type='dataframe')
y.values.ravel()
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Null hypothesis rejected, proportion of black in the town has a non 0 betacoefficient 
# Null hypothesis
```

```python
# if want to use explore every single feature, use combinations, then choose the best model
lm = LinearRegression()
for n in range(5,11):
    kf_shuffle = KFold(n_splits=n, shuffle=True)
    scores = cross_val_score(lm, X, y, cv=kf_shuffle)
    print(scores.mean())
```
