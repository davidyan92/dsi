```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns

unique_id = [x for x in range(1,101)]
df = pd.read_csv('pokedex_basic.csv', nrows=100)
del df['PokedexNumber']
df = df.reindex(unique_id)
print(len(df))
df.head()
```

```python
# scatter plot using seaborn
sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
# Scatter plot with more arguments
# 'fit_reg' toggles best fit regression line
# 'hue' enables 3rd dimension display of information
sns.lmplot(x='Attack', y='Defense', data=df, 
           fit_reg=False)
```

```python
# Boxplot
# resizing chart
stats_df = df.drop(['Total'], axis=1)
fig, ax = plt.subplots(figsize=(20,6))
sns.boxplot(data=stats_df, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(20,6))
plt.xlabel('Attributes')
plt.ylabel('Value')
sns.set_style('whitegrid')
sns.violinplot(data=stats_df, ax=ax)
```
