```python
def fill_zero_values(df_series):
    total_abv_0 = df_series.where(df_series > 0).value_counts().sum()
    weight_df = (df_series.where(df_series > 0).value_counts() / total_abv_0).to_frame()
    values = weight_df.index.tolist()
    weights = weight_df.values.ravel()
    return values, weights
```

```python
# LotFrontage, replace nan with weighted random filling
values, weights = fill_zero_values(house.LotFrontage)
house.LotFrontage.fillna(np.random.choice(values, p=weights), inplace=True)
house['LotFrontage'] = house['LotFrontage'].map(lambda x: np.random.choice(values, p=weights) if x == 0. else x)
```

```python
# Convert Lot Area into Percentile, cause we have some very large values
house['LotArea_pctl'] = house['LotArea'].map(lambda x: stats.percentileofscore(house.LotArea.values, x, kind='mean'))
```

```python
# Divide year built into groups of 4, 1872 - 1900, 1901 - 1950, 1951 - 2000, 2001-2010 (reference years)
# Engineer Year remodel into Remodelled or not, remodel year > built year = 1 else 0
house['YearBuilt_1872_1900'] = house['YearBuilt'].map(lambda x: 1 if x >= 1872 and x <= 1900 else 0)
house['YearBuilt_1901_1950'] = house['YearBuilt'].map(lambda x: 1 if x >= 1901 and x <= 1950 else 0)
house['YearBuilt_1951_2000'] = house['YearBuilt'].map(lambda x: 1 if x >= 1951 and x <= 2000 else 0)
house['Remod_yes'] = house['YearRemodAdd'] > house['YearBuilt']
house['Remod_yes'] = house['Remod_yes'].map(lambda x: 1 if x == True else 0)
# To handle nan values in Garage year built, lets use corresponding year built as replacement
house.GarageYrBlt.fillna(house.YearBuilt, inplace=True)
#  Engineer Garage year same way as Year Built with reference to 2001-2010
house['GarageYrBlt_1872_1900'] = house['GarageYrBlt'].map(lambda x: 1 if x >= 1872 and x <= 1900 else 0)
house['GarageYrBlt_1901_1950'] = house['GarageYrBlt'].map(lambda x: 1 if x >= 1901 and x <= 1950 else 0)
house['GarageYrBlt_1951_2000'] = house['GarageYrBlt'].map(lambda x: 1 if x >= 1951 and x <= 2000 else 0)
```

# Optimization & Model Selection

```python
# Deal with multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif["vif_factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
```

```python
indep_var_mask = vif[vif['vif_factor']<=5].features.values.ravel()
```

```python
# Only deal with independent variables
Y_train = y_train.values.ravel()
Y_test = y_test.values.ravel()
Xs_train = StandardScaler().fit_transform(X_train[indep_var_mask])
Xs_test = StandardScaler().fit_transform(X_test[indep_var_mask])
```

```python
# Compute Plain Vanilla Linear regression
linreg = LinearRegression()

linreg_scores = cross_val_score(linreg, Xs_train, Y_train, cv=10)

print linreg_scores
print np.mean(linreg_scores)
```

```python
# Elastic Net
l1_ratios = np.linspace(0.01, 1.0, 25)
optimal_enet = ElasticNetCV(l1_ratio=l1_ratios, n_alphas=100, cv=10,
                            verbose=1)
optimal_enet.fit(Xs_train, Y_train)
print optimal_enet.alpha_
print optimal_enet.l1_ratio_
```

```python
# Pure Lasso from Elastic Net
enet = ElasticNet(alpha=optimal_enet.alpha_, l1_ratio=optimal_enet.l1_ratio_)
enet_scores = cross_val_score(enet, Xs_train, Y_train, cv=10)
print enet_scores
print np.mean(enet_scores)
```

```python
# Get all the Non Zero Betas
lasso_coefs = pd.DataFrame({'pred':X_train[indep_var_mask].columns,
                            'coef':optimal_enet.coef_,
                            'abs_coef':np.abs(optimal_enet.coef_)})

lasso_coefs.sort_values('abs_coef', inplace=True, ascending=False)
```

```python
Nonzerobeta_features = lasso_coefs[lasso_coefs.abs_coef != 0]
```
